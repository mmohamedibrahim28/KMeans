```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv("/Users/ibu/Downloads/players_22.csv")
```

```python
features = ["overall","potential","wage_eur", "value_eur","age"]
```

```python
players=players.dropna(subset=features)
```

```python
data=players[features].copy()
```

```python
data
```

```python
# 1. Scale the data
# 2. Initialize the random centroids
# 3. Label each data point
# 4. Update our centroids
# 5. Repeat 3,4 until centroids stop changing
```

```python
# 1. Scale the data
data = (data-data.min())/(data.max()-data.min())*9 + 1
```

```python
data.describe()
```

```python
data.head()
```

```python
# 2. Initialize the random centroids
def random_centroids(data,k):
    centroids=[]
    for i in range(k):
        centroid = data.apply(lambda x: float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids,axis=1)
```

```python
centroids = random_centroids(data,5)
```

```python
centroids
```

```python
# 3. Label each data point
def get_labels(data,centroids):
    distances = centroids.apply(lambda x: np.sqrt (((data-x)**2).sum(axis=1))) #using geometric distance formula to calulate distance between each data point and random centroids
    return distances.idxmin(axis=1) #using the min index number for each player
```

```python
distances #for min index - for player 0 is 2 which is the min distance from centroid to data point
```

```python
labels=get_labels(data,centroids)
```

```python
labels
```

```python
labels.value_counts() # now we have the clusters for each players
```

```python
# 4. Update our centroids
# we will get the geometric mean of the data points in the cluster and update the centroid point
```

```python
#geometric mean
data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean()))
```

```python
 data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T #making the output similar to the random centroid 
```

```python
def new_centroids(data,labels,k):
     return data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output 
```

```python
#for Ploting
def plot_clusters(data,labels,centroids,iteration):
    pca = PCA(n_components=2)
    data_2d= pca.fit_transform(data)
    centroids_2d= pca.fit_transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration{iteration}')
    plt.scatter(x=data_2d[:,0],y=data_2d[:,1],c=labels)

    plt.scatter(x=centroids_2d[:,0],y=centroids_2d[:,1])
    plt.show()
    
```

```python
# 5. Repeat 3,4 until centroids stop changing with iteration =100
#getting the pieces together
max_iterations =100
k=5
centroids = random_centroids(data,k)
old_centroids = pd.DataFrame()
iteration = 1

while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids=centroids
    
    labels = get_labels(data, centroids)
    centroids = new_centroids(data,labels, k)
    plot_clusters(data,labels,centroids,iteration)
    iteration +=1
```
